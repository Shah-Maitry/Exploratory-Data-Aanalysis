```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#line chart
x=[1,2,3,4,5,6,7,8,9]
y=[12, 45, 56, 89, 13, 35, 71, 20, 6]
z=[92,45,23,16,89,43,78,56,99]
plt.xlabel('x-axis')
plt.ylabel('random')
plt.title('haahaha')
plt.plot(x,y, label= 'y')
plt.plot(x,z, label= 'z')
plt.legend(loc="upper left", fontsize='large')
plt.grid()
```

```python
#barh chart
import numpy as np
x=[1,2,3,4,5,6,7,8,9]
y=[12, 45, 56, 89, 13, 35, 71, 20, 6]
z=[92,45,23,16,89,43,78,56,99]
xpos=np.arange(1,len(x)+1)
#plt.xticks(xpos, y)
ypos=np.arange(1,len(y)+1)
plt.yticks(ypos, x)
plt.xlabel('x-axis')
plt.ylabel('random')
plt.title('haahaha')
plt.barh(x,y, label='y', color=['yellow'])
plt.legend(loc='lower right')
plt.grid()
```

```python
#barchart
import numpy as np
x=[1,2,3,4,5,6,7,8,9]
y=[12, 45, 56, 89, 13, 35, 71, 20, 6]
z=[92, 45, 23, 16, 89, 43, 78, 56, 99]
xpos=np.arange(1,len(x)+1)
plt.xticks(xpos, y)
plt.xlabel('x-axis')
plt.ylabel('random')
plt.title('haahaha')
plt.xticks(xpos, x)
#plt.yticks(ypos, y)
plt.bar(xpos+0.2,y, label='y',width=0.4, color=['black'])
plt.bar(xpos-0.2,z, label='z',width=0.4, color=['green'])
plt.legend()
plt.grid()
```

```python
#histogram
y=[12, 45, 56, 89, 13, 35, 71, 20, 6]
mn=min(y)
mx=max(y)
plt.xlabel('x-axis')
plt.ylabel('random')
plt.title('haahaha')
plt.hist([y,z], rwidth=.80, label=['y', 'z'], color=['blue', 'red'])
plt.legend()
```

```python
plt.xlabel('x-axis')
plt.ylabel('random')
plt.title('haahaha')
plt.hist([y,z], rwidth=.80, label=['y', 'z'], color=['blue', 'red'], histtype='step')
plt.legend()
```

```python
mn=min(y)
mx=max(y)
plt.xlabel('x-axis')
plt.ylabel('random')
plt.title('haahaha')
plt.hist([y,z], bins=[0,30,60,mx], rwidth=.80, color=['orange', 'green'], label=['y', 'z'], orientation='horizontal')
plt.legend()
plt.grid()
```

```python
#piechart
x=[1,2,3,4,5,6,7,8,9]
y=[12, 45, 56, 89, 13, 35, 71, 20, 6]
z=[92, 45, 23, 16, 89, 43, 78, 56, 99]
plt.axis('equal')
plt.pie(y,labels=x, autopct='%2.1f%%',radius=1.3,explode=[0,0,0.2,0,0,0,0,0,0], startangle=90)
plt.show()
```

```python
plt.pie(z,labels=x, autopct='%2.1f%%',radius=1.4,explode=[0,0,0.2,0,0,0,0,0,0], startangle=270)
plt.show()
```

```python
plt.pie(z,labels=x, autopct='%2.1f%%',radius=1.4,explode=[0,0,0.2,0,0,0,0,0,0], startangle=270)
plt.savefig('pichart.png',facecolor='red',edgecolor='b', pad_inches=3)
```
