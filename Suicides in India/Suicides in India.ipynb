# Problem 1

   ## Which Year has a Max And Min Suicide Cases in India ?? 

###   In This Problem we have to find Year wises Max and Min cases. Therefore we Using Groupby to solve this problem and find Maxium cases of India & Min cases of India using their syntax.    

```python
cases_df = df_sui[['Year','Total']]
```

```python
cases_df
```

```python
a = cases_df.groupby(['Year']).sum()
```

```python
a
```

```python
print(cases_df['Total'].sum())
```

```python
a.reset_index()
```

```python
print(a[a.Total == a.Total.max()])
```

```python
print(a[a.Total == a.Total.min()])
```

```python
a.describe()
```

### Thus, Here Maxium case of suicides are 1219499 which is in Year 2011 & Minimum Case of suicides are 976464 which is in Year 2001 . 


### Hear We also plot graph -- Year Vs Total

```python
sns.lineplot('Year', 'Total',data = df_sui, ci=None)
plt.show()
```

#  Some Different Graphs In Different Style (Type)




# Graph 1  --  Lineplot--  Year Vs Total

```python
sns.lineplot('Year','Total',hue='Gender',data = df_sui, ci=None)
plt.show()
```

### This Graph Type is Line plot Which is a very simple and very normal type. This type of graph are mostly use in Stock Market like sensex graph, Gdp Graph etc... 

# Graph 2  --  Heatmap  --   Year Vs State

```python
sy = df_sui[['State','Year']]
```

```python
sy
```

```python
df = df_sui[~df_sui['State'].isin(['Total (All India)','Total (States)','Total (Uts)'])]
```

```python
sy = df.pivot_table(index='State',columns='Year',values='Total')
```

```python
sy
```

```python
sns.heatmap(sy)
plt.show()
```

### This type is Heatmap which used to show relationship between 2 variables, One ploted on each axis.By Observing how cell colour change across each axis, we can observe if there are any patterns in value for one or both variables. This type of graph are mostly use in Online Business like Finance,OTT Media etc...

# Graph 5 -- FacetGraph  --  Year Vs Total Vs Type

```python
sns.relplot('Year','Total',hue = 'Type',data = df_sui,col = 'State',col_wrap = 3)
plt.draw()
```

### This Type of Graph is Known as Facetgraph which useful approch to explore medium dimensional data, is by Drawing multiple instances of the same plot on different subset of our DataSet.
