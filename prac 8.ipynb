```python
''' Write a pandas programme to select the rows where the no. of attempts in the examination is grater than 2 '''
import pandas as pd
import numpy as np

exam_data={
    'name': ['Anastssia', 'Dima', 'Katherine', 'James', 'Emaily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
    'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
    'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no','no', 'yes']
    }
labels=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j',]
exam= pd.DataFrame(exam_data, index=labels)
print('exam_data')
print()
print(exam)
print()
print('The rows where the no. of attempts in the examination is grater than 2')
print()
print(exam.loc[exam.attempts>2])
```

```python
''' Write a pandas programme to sort dataframe first by "name" in descending order and than "score" in aescending order'''
import pandas as pd
import numpy as np

exam_data={
    'name': ['Anastssia', 'Dima', 'Katherine', 'James', 'Emaily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
    'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
    'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no','no', 'yes']
    }
labels=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j',]
exam= pd.DataFrame(exam_data, index=labels)
print('exam_data')
print()
print(exam)
print()
print('sorted dataframe')
print()
print(exam.sort_values(by=['name', 'score'], ascending=[False, True]))
```

```python
''' Write a pandas programme to replace "yes" with "True" and "no" with "False"'''
import pandas as pd
import numpy as np

exam_data={
    'name': ['Anastssia', 'Dima', 'Katherine', 'James', 'Emaily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
    'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
    'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no','no', 'yes']
    }
labels=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j',]
exam= pd.DataFrame(exam_data, index=labels)
print('exam_data')
print()
print(exam)
print()
print('After replacing "yes" with "True" and "no" with "False"')
exam.qualify=exam.qualify.map({'yes': True, 'no': False})
print()
print(exam)
```
